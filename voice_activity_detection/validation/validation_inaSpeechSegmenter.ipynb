# References

- [GitHub Repo](https://github.com/ina-foss/inaSpeechSegmenter)
- [Paper](https://www.researchgate.net/publication/324752429_An_Open-Source_Speaker_Gender_Detection_Framework_for_Monitoring_Gender_Equality) : An Open-Source Speaker Gender Detection Framework for Monitoring Gender Equality
    - Accepted in Conference: 2018 IEEE International Conference on Acoustics, Speech, and Signal Processing (ICASSP 2018)At: Calgary, Alberta, Canada
- [Copy of the Paper in my References](https://drive.google.com/file/d/1NnsSY-zlBmp6fUYoI3dxxygwwRc4o55T/view?usp=sharing)
- **Idea**: 
  - Gender detection systems based on **Gaussian Mixture Models**, **i-vectors** and **Convolutional Neural Networks (CNN)** were trained using an internal database of 2,284 French speakers and evaluated using REPERE challenge corpus.
  - In the present work, we design an automatic speaker gender detection system for large scale monitoring of raw audiovisual streams in order to reduce the bias associated to a particular measure context.

- **Criticalities highlighted by the authors**: 
  - The robustness of speech gender detection systems is still challenged by speakers having marked accents (regional, foreign), extreme pitch ranges, or speaking using non-standard intonation.

- **Data** used to **train** the system: 
  - Audiovisual streams contain speech recorded in various conditions (telephone of variable quality, studio, spontaneous and prepared speech, speech-over-music, speech-over-noise) which should be handled robustly. 32000 speech excerpts, corresponding to 1780 male (94 hours) and 494 female speakers (27h).
  - **Summary of dataset**: internal dataset of **2284** speakers, male speakers = 1780, female speakers = 494

- **Architecture**: 
Gender detection processing pipeline is composed of several modules. 
  1. A Speech/Music segmenter based on CNN is used to discard music and empty segments. References for the CNN used here: 
    - Paper: [INA’S MIREX 2018 MUSIC AND SPEECH DETECTION SYSTEM](https://drive.google.com/file/d/1hQBy1ZjeAGRcSX9idbx4PNIi3OxszsKc/view?usp=sharing)
    - Paper: [Investigating the Use of Semi-Supervised Convolutional
Neural Network Models for Speech/Music Classification
and Segmentation](https://drive.google.com/file/d/122tdo19qje7q9gCom0udPVyivX0ic778/view?usp=sharing)

  2. Features corresponding to speech segments are then computed using a common extraction framework. Feature extraction was realized on audio excerpts sampled at **16KHz** using SIDEKIT.
    - 2 Acoustic Features: 
        - 24 **Mel-scaled filter-banks coefficients**
        - **MFCCs** obtained by applying Discrete Cosine Transform (DCT) to those filterbank coefficients to extract **19 Mel-Frequency Cepstrums coefficients (MFCC)**, which were augmented with log energy.
          - Time derivatives (∆) and acceleration (∆∆) coefficients were concatenated to MFCCframes, resulting in 60-dimensional input features used by GMM and i-vector models.
  3. A simple **energy threshold** is used to discard frames associated with low energy. 
  4. GMM, i-vector and CNN systems are then used to classify the remaining speech segments into male and female excerpts. **Best Results were obtained with the CNN model**. 

*From what I have understood from the Papers* the CNN architecture used for Speech-Music detection and Gender-Classification is the same. What is different is the kind of features that are fed to the network. 

- **Strength points**: 
  - Since the model is trained with a particular interest on the gender balance, we have not to worry about the fact that male voice could be better recognize than female voice. This is important because in the context of **5GMED** we need to be able to distinguish speech without making distinctions on the gender of the person. 
  - The model is first built to distinguish speech from noise and/or music and then to recognize the gender of the speaker. Since the model is capable to recognize the gender,**we expect that is also quite good and precise in distinguish speech from noise**. 
  - **Connection with Speech Emotion Recogniton (from the paper)**: More recent studies have used speaker gender detection as a preprocessing step for selecting gender specific emotion recognition engines. Ref [paper Modeling Gender Information for Emotion Recognition Using Denoising Autoencoders.](https://mediatum.ub.tum.de/doc/1238169/1238169.pdf)  

- **Notes**: 
  - The architecture is not available. The parts available regard: 
    - The weights of the full model 
    - The utils for feature extraction (SIDEKIT for mfcc and the segmenter)

## Dataset Utils

```python
def extract_ravdess(dataset_path=None):
    if dataset_path is None:
        RAV = "datasets/ravdess-emotional-speech-audio/audio_speech_actors_01-24/"
    else:
        RAV = dataset_path

    dir_list = os.listdir(RAV)
    dir_list.sort()

    emotion = []
    gender = []
    path = []
    for i in dir_list:
        #print(i)
        if not i.startswith('.'): #exclude .DS_Store files
            fname = os.listdir(RAV + i)

            for f in fname:
                if not f.startswith('.'): #exclude .DS_Store files
                    part = f.split('.')[0].split('-')
                    emotion.append(int(part[2]))
                    temp = int(part[6])
                    if temp % 2 == 0:
                        temp = "female"
                    else:
                        temp = "male"
                    gender.append(temp)
                    path.append(RAV + i + '/' + f)

    RAV_df = pd.DataFrame(emotion)
    RAV_df = RAV_df.replace(
        {1: 'speech', 2: 'speech', 3: 'speech', 4: 'speech', 5: 'speech', 6: 'speech', 7: 'speech', 8: 'speech'})

    RAV_df = pd.concat([pd.DataFrame(gender), RAV_df], axis=1)
    RAV_df.columns = ['gender', 'emotion']
    RAV_df['labels'] = RAV_df.emotion
    RAV_df['source'] = 'RAVDESS'
    RAV_df = pd.concat([RAV_df, pd.DataFrame(path, columns=['path'])], axis=1)
    RAV_df = RAV_df.drop(['gender', 'emotion'], axis=1)
    RAV_df.labels.value_counts()

    return RAV_df

def extract_tess(dataset_path=None):
    if dataset_path is None:
        TESS = "datasets/tess/TESS Toronto emotional speech set data/"
    else:
        TESS = dataset_path


    tess_directory_list = os.listdir(TESS)

    file_emotion = []
    file_path = []

    for dir in tess_directory_list:
        if not dir.startswith('.'):
            directories = os.listdir(TESS + dir)
            for file in directories:
                if not file.startswith('.'):
                    file_emotion.append('speech')
                    #print(TESS + dir + '/' + file)
                    file_path.append(TESS + dir + '/' + file)

    # dataframe for emotion of files
    emotion_df = pd.DataFrame(file_emotion, columns=['labels'])
    emotion_df['source'] = 'TESS'
    # dataframe for path of files.
    path_df = pd.DataFrame(file_path, columns=['path'])
    TESS_df = pd.concat([emotion_df, path_df], axis=1)
    TESS_df.head()

    TESS_df['labels'].unique()

    return TESS_df


def extract_emodb(dataset_path=None):
    if dataset_path is None:
        EMODB = "datasets/emo-db/wav/"
    else:
        EMODB = dataset_path

    #emodb_directory_list = os.listdir(EMODB)
    emotion = []
    path = []

    for root, dirs, files in os.walk(EMODB):
        for name in files:
            if not name.startswith('.'):
                #print("NAME ", name)
                if name[0:2] in '0310111215':  # MALE
                    if name[5] == 'W':  # Ärger (Wut) -> Angry
                        emotion.append('speech')
                    elif name[5] == 'L':  # Langeweile -> Boredom
                        emotion.append('speech')
                    elif name[5] == 'E':  # Ekel -> Disgusted
                        emotion.append('speech')
                    elif name[5] == 'A':  # Angst -> Angry
                        emotion.append('speech')
                    elif name[5] == 'F':  # Freude -> Happiness
                        emotion.append('speech')
                    elif name[5] == 'T':  # Trauer -> Sadness
                        emotion.append('speech')
                    else:
                        emotion.append('unknown')
                else:
                    if name[5] == 'W':  # Ärger (Wut) -> Angry
                        emotion.append('speech')
                    elif name[5] == 'L':  # Langeweile -> Boredom
                        emotion.append('speech')
                    elif name[5] == 'E':  # Ekel -> Disgusted
                        emotion.append('speech')
                    elif name[5] == 'A':  # Angst -> Angry
                        emotion.append('speech')
                    elif name[5] == 'F':  # Freude -> Happiness
                        emotion.append('speech')
                    elif name[5] == 'T':  # Trauer -> Sadness
                        emotion.append('speech')
                    else:
                        emotion.append('unknown')

                path.append(os.path.join(EMODB, name))

    emodb_df = pd.DataFrame(emotion, columns=['labels'])
    # remove 'unknown'rows
    emodb_df.drop(index=emodb_df[emodb_df['labels'] == 'unknown'].index, inplace=True)
    emodb_df['source'] = 'EMODB'
    emodb_df = pd.concat([emodb_df, pd.DataFrame(path, columns=['path'])], axis=1)

    return emodb_df

def extract_extra_sounds_speech():
    SPEECH_PATH = "datasets/recorded_sounds/speech/"
    labels = []
    path = []

    for root, dirs, files in os.walk(SPEECH_PATH):
        for name in files:
            if not name.startswith('.'):
                labels.append('speech')
                path.append(os.path.join(SPEECH_PATH, name))

    speech_df = pd.DataFrame(labels, columns=['labels'])
    speech_df['source'] = 'recorded_speech'
    speech_df = pd.concat([speech_df, pd.DataFrame(path, columns=['path'])], axis=1)
    return speech_df

def extract_extra_sounds_noise():
    NOISE_PATH = "datasets/recorded_sounds/other/"
    labels = []
    path = []

    for root, dirs, files in os.walk(NOISE_PATH):
      for name in files:
          if not name.startswith('.'):
              labels.append('noise')
              path.append(os.path.join(NOISE_PATH, name))

    noise_df = pd.DataFrame(labels, columns=['labels'])
    noise_df['source'] = 'recorded_speech'
    noise_df = pd.concat([noise_df, pd.DataFrame(path, columns=['path'])], axis=1)
    return noise_df

def datasets_union():
    rav_df = extract_ravdess()
    tess_df = extract_tess()
    emo_df = extract_emodb()
    speech_df = extract_extra_sounds_speech()
    noise_df = extract_extra_sounds_noise()
    df = pd.concat([rav_df, tess_df, emo_df, speech_df, noise_df], axis=0)
    df.to_csv("datasets/validation_datasets.csv", index=False)
    return df

def noise(data):
    noise_amp = 0.035*np.random.uniform()*np.amax(data)
    data = data + noise_amp*np.random.normal(size=data.shape[0])
    return data

def add_random_noise(df=None):
    '''
    Write copies of the audio files by adding random noise
    :return:
    '''
    if df is None:
        df = datasets_union()
    else:
        df = pd.read_csv('datasets/validation_datasets.csv')

    counter = 0
    rows_list = []
    for index, row in tqdm(df.iterrows(), total=df.shape[0]):
        filepath = row['path']
        label = row['labels']
        source = row['source']
        if label == 'speech':
            provv_filename = filepath.replace(".wav", "")
            new_file_name = provv_filename + "_noise.wav"
            data, sample_rate = librosa.load(filepath)
            # noise injection
            x = noise(data)
            sf.write(new_file_name, x, sample_rate)
            dict_row = {"labels": label, "source": source, "path": filepath}
            rows_list.append(dict_row)
            counter += 1
    df_noise_files = pd.DataFrame(rows_list)
    concat = pd.concat([df, df_noise_files])
    concat.reset_index(inplace = True, drop = True)
    print('Done!')
    return concat
"""
def add_ambient_noise():
    '''
    Write copies of the audio files by adding train and crowd noise
    :return:
    '''
"""



```

# Segmentation

## Segmentation Utils

```python
def analyze_segmentation(d):
  count_speech = 0
  for el in d:
    if el[0] == 'speech':
      count_speech+=1
  return count_speech

def segmentation(df, segmenter):
  results = []
  counter = 0
  for index, row in tqdm(df.iterrows()):
    #print(counter)
    filepath = row['path']

    count_speech = analyze_segmentation(segmenter(filepath))
    if count_speech > 0:
      results.append("speech")
    else:
      results.append("noise")
    print(results[counter])
    counter += 1

  df['results'] = results

  return df

def save_results(df):
  with open('datasets/dataset_VAD_results.pickle', 'wb') as handle:
    pickle.dump(df, handle, protocol=pickle.HIGHEST_PROTOCOL)

def load_results():
  with open('datasets/dataset_VAD_results.pickle', 'rb') as handle:
    df = pickle.load(handle)
    return df

def validation(df):
  """
  :param df_res: df_res
  :return: Percentage of misclassified files
  """
  differences = df[df['labels'] != df['results']]
  counter_noise = 0
  misclassified_speech = 0
  misclassified_noise = 0

  for index, row in differences.iterrows():
    filepath = row['path']
    true_label = row['labels']
    predict_label = row['results']
    if "noise" in filepath:
      counter_noise += 1
    if true_label == 'speech' and predict_label == 'noise':
      misclassified_speech += 1
    elif true_label == 'noise' and predict_label == 'speech':
      misclassified_noise += 1

  perc_misc = (100 * differences.shape[0]) / df.shape[0]
  print("Percentage of misclassified files: ", perc_misc, "%")

  print("Among the misclassified files there are ", counter_noise, "files on which we injected noise")
  print("Among the misclassified files, ", misclassified_speech, " files of speech were labelled as noise")
  print("Among the misclassified files, ", misclassified_noise, " files of noise were labelled as speech")

def analyze_dataset_misclassified(paths): 
    # check how many misclassified files refer to RAVDESS dataset 
    count_rad = 0
    for i in paths: 
      if 'ravdess' in i: 
        count_rad +=1 
    print("Among the misclassified files there are ", count_rad, "files belonging to RAVDESS dataset")
          

def analyze_gender_misclassified(paths):
    # check how many females and males are in misclassified radvess 
    count_rad = 0
    females = []
    males = []
    for f in paths: 
      if 'ravdess' in f: 
        part = f.split('.')[0].split('-')
        temp = int(part[10])
        if temp%2 == 0:
            females.append("female")
        else:
            females.append("male")
    print("Males: ", len(males))
    print("Females: ", len(females))

def analyze_actresses(paths):
    # Check how many different females are present in misclassified ravdess files 
    females_dictionary = {}
    for f in paths: 
      if 'ravdess' in f: 
        part = f.split('.')[0].split('-')
        temp = int(part[10])
        if temp%2 == 0:
          if temp not in females_dictionary.keys():
            females_dictionary[temp] = 1
          else: 
            females_dictionary[temp] += 1

    print(females_dictionary)

def misclassified_emotions_per_actress(paths): 
    females_dictionary = {}
    for f in paths: 
      if 'ravdess' in f: 
        part = f.split('.')[0].split('-')
        emotion = int(part[6])
        if emotion == 1: 
          emotion = 'neutral'
        elif emotion == 2:
          emotion = 'calm'
        elif emotion == 3: 
          emotion = 'happy'
        elif emotion == 4:
          emotion = 'sad'
        elif emotion == 5: 
          emotion = 'angry'
        elif emotion ==6: 
          emotion = 'fear'
        elif emotion == 7: 
          emotion = 'disgust'
        else:
          emotion = 'surprise'

        temp = int(part[10])
        if temp%2 == 0:
          if temp not in females_dictionary.keys():
            females_dictionary[temp] = emotion 
          else: 
            em = females_dictionary.get(temp)
            females_dictionary[temp] = em + ',' + emotion

    for k,v in females_dictionary.items(): 
      print(k, v)
```
